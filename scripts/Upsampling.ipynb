```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import scale
from sklearn.utils import resample
```

```python
train_data = pd.read_csv("train_dataset3.csv", index_col=0)
```

```python
train_data = train_data.drop(columns=["Unnamed: 0.1"])
```

```python
train_data
```

* Label encode `PotentialFraud`

```python
train_data['PotentialFraud'] = train_data['PotentialFraud'].apply(lambda x : 1 if x=='Yes' else 0)
```

```python
# Check the numbers for the response variable
train_data['PotentialFraud'].value_counts()
```

## Upsample the minority class via SMOTE

```python
minorIndex = train_data[train_data['PotentialFraud']==1].Provider.values

resampled = resample(minorIndex, n_samples=4904-506)
LEFT = pd.DataFrame(list(train_data.Provider) + list(resampled), columns=['INDEX'])

train_data = pd.merge(LEFT, train_data, how='left', left_on ='INDEX', right_on='Provider')
```

```python
train_data['PotentialFraud'].value_counts()
```

```python
train_data.to_csv("train_sampled.csv")
```
